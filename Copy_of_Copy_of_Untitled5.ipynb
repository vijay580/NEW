```
x_train = np.array(x_train)
print(x_train.shape)
y_train = np.array(y_train)
x_val = np.array(x_val)
y_val = np.array(y_val)
for i in range(0,25):
  print(encoded[i])
```

```
def train_mnist_conv(dense=512,dropout=0.4,activation='relu',eta=0.00146,maps=96,ksize=7,norm=2):
   
    class myCallback(tf.keras.callbacks.Callback):
        def on_epoch_end(self, epoch, logs={}):
            if(logs.get('acc')>0.998):
              #print("\nReached 99.8% accuracy so cancelling training!")
              self.model.stop_training = True

    
    callbacks = myCallback()
    model = tf.keras.models.Sequential([
            
            tf.keras.layers.Conv2D(maps, (ksize), activation=activation, input_shape=(128, 128, 3)),
            tf.keras.layers.BatchNormalization(),
        
            tf.keras.layers.Conv2D(maps, (ksize), activation=activation),
            tf.keras.layers.BatchNormalization(),
        
            tf.keras.layers.Conv2D(maps, (ksize+2),strides=2,padding='same', activation=activation),
            tf.keras.layers.BatchNormalization(),
            tf.keras.layers.Dropout(dropout),
        
            tf.keras.layers.Conv2D((maps*2), (ksize), activation=activation),
            tf.keras.layers.BatchNormalization(),
        
            tf.keras.layers.Conv2D((maps*2), (ksize), activation=activation),
            tf.keras.layers.BatchNormalization(),
        
            tf.keras.layers.Conv2D((maps*2), (ksize+2),strides=5,padding='same', activation=activation),
            tf.keras.layers.BatchNormalization(),
            tf.keras.layers.Dropout(dropout),
        
            tf.keras.layers.Flatten(),
        
            tf.keras.layers.Dense(dense, activation=activation,kernel_constraint=tf.keras.constraints.max_norm(norm)),
            tf.keras.layers.BatchNormalization(),
            tf.keras.layers.Dropout(dropout),
            tf.keras.layers.Dense(dense, activation=activation,kernel_constraint=tf.keras.constraints.max_norm(norm)),
        
            tf.keras.layers.BatchNormalization(),
            tf.keras.layers.Dropout(dropout),
        
            tf.keras.layers.Dense(5, activation='softmax')
    ])
    opt = tf.keras.optimizers.Adam(learning_rate=eta)
    model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
    #model.summary()
    history = model.fit(
        x_train, y_train, epochs=36, #callbacks=[callbacks]
        validation_data=(x_val,y_val)
    )
    model.save(base_dir+'weights')
    if activation == 'relu':
       y_pred = model.predict(images)

    import pandas as pd
  #print(y_pred)
    for i, d in enumerate(y_pred):
        y_pred[i] = y_pred[i] // max(d)

    y_pred = np.array(y_pred, dtype=np.uint8)
# save data into csv file
#for i in y_onehot:
    save = pd.DataFrame(y_pred, columns=[0, 1, 2, 3, 4])
    save.to_csv(path+'new.csv', index=False, header=False) 
    f.write("\n --------------------new--------------------")
  #(dense=128,dropout=0.4,activation='relu',eta=0.001,maps=32,ksize=3)
  

    f.write("\n Denselayer: ")
    f.write(str(dense))
    f.write("\n Dropout: ")
    f.write(str(dropout))
    f.write("\n Activation: ")
    f.write(str(activation))
    f.write("\n Learning Rate: ")
    f.write(str(eta))
    f.write("\n Maps: ")
    f.write(str(maps))
    f.write("\n kernel size: ")
    f.write(str(ksize))
   
    
    f.write("\n Training_Loss: ")
    f.write(str(history.history['loss'][-1]))
    f.write("\n Training_acc: ")
    f.write(str(history.history['acc'][-1]))
    f.write("\n Valid_Loss: ")
    f.write(str(history.history['val_loss'][-1]))
    f.write("\n Valid_acc: ")
    f.write(str(history.history['val_acc'][-1]))

    f.write("\n Training_Loss Complete: ")
    f.write(str(history.history['loss']))
    f.write("\n Training_acc Complete: ")
    f.write(str(history.history['acc']))
    f.write("\n Valid_Loss Complete: ")
    f.write(str(history.history['val_loss']))
    f.write("\n Valid_acc Complete: ")
    f.write(str(history.history['val_acc']))
    
    
    return history.epoch, history.history['acc'], history.history['val_acc'],history.history['loss'],history.history['val_loss']
```
